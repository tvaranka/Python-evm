```python
import py_evm
import matplotlib.pyplot as plt
```

```python
video = py_evm.load_video("data/baby.mp4")
video.shape
```

```python
plt.figure(figsize=(16, 8))
plt.imshow(video[210])
plt.show()
```

```python
# mm stands for motion magnified
mm_video = py_evm.magnify(video)
mm_video.shape
```

```python
plt.figure(figsize=(16, 8))
plt.imshow(mm_video[210])
plt.show()
```

```python
py_evm.save_video(mm_video, "mm_baby.mp4")
```
